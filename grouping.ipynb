```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/gdrive/My Drive/data/StudentsPerformance.csv")
```

```python
df.head()
```

```python
#average scores in relation to lunch type
reduced_students = df.loc[df['lunch']=="free/reduced"]
print('reduced')
print('mean math = ', reduced_students['math score'].mean())
print('mean reading = ', reduced_students['reading score'].mean())
print('mean writing = ', reduced_students['writing score'].mean())
print()
print("variance math = ", reduced_students["math score"].var())
print("variance reading = ", reduced_students["reading score"].var())
print('variance = ', reduced_students['writing score'].var())
```
