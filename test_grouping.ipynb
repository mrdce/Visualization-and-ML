```python
import pandas as pd

my_stat = pd.read_csv('/gdrive/My Drive/data/my_stat.csv')
my_stat
```

```python
subset_1 = my_stat.head(10).iloc[:, [0, 2]]
subset_2 = my_stat.drop([0, 4]).iloc[:, [1, 3]]
#10 first rows, columns 1 and 3; all rows but 1 and 5, columns 2 and 4
```

```python
#В переменную subset_1 сохраните только те наблюдения, у которых значения переменной V1  строго больше 0, и значение переменной V3  равняется 'A'.
subset_1 = my_stat.loc[my_stat.V1 > 0].loc[my_stat.V3 == 'A']
#В переменную  subset_2  сохраните только те наблюдения, у которых значения переменной V2  не равняются 10, или значения переменной V4 больше или равно 1.
subset_2 = my_stat.query('V2 != 10 | V4 >= 1')
```

```python
#В этих данных (my_stat) создайте две новые переменных: 1) V5 = V1 + V4 2) V6 = натуральный логарифм переменной V2
import numpy as np
my_stat['V5'] = my_stat['V1'] + my_stat['V4']
my_stat['V6'] = np.log(my_stat['V2'])
```

```python
#Переименуйте колонки в данных
my_stat = my_stat.rename(columns = {'V1': 'session_value',
                          'V2': 'group',
                          'V3': 'time',
                          'V4': 'n_users'})
my_stat
```

```python
my_stat = pd.read_csv('/gdrive/My Drive/data/my_stat_1.csv')
# В переменной session_value замените все пропущенные значения на нули.
my_stat['session_value'] = my_stat['session_value'].fillna(value=0)
# В переменной n_users замените все отрицательные значения на медианное значение переменной n_users (без учета отрицательных значений, разумеется).
my_stat.loc[my_stat.n_users < 0.0, 'n_users'] = my_stat.loc[my_stat.n_users >= 0, 'n_users'].median()
my_stat
```

```python
# Pассчитайте среднее значение переменной session_value для каждой группы (переменная group), в получившемся dataframe  переменная group не должна превратиться в индекс.
# Также переименуйте колонку со средним значением session_value в mean_session_value. Получившийся результат сохраните в dataframe с именем mean_session_value_data.
mean_session_value_data = my_stat.groupby('group', as_index=False).agg({'session_value': 'mean'})
mean_session_value_data = mean_session_value_data.rename(columns = {'session_value': 'mean_session_value'})
mean_session_value_data
```
