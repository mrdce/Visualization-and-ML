Анализ распределения параметров ирисов.

```python
df = pd.read_csv('/gdrive/My Drive/data/iris.csv')
df.head()
```

```python
for column in df:
    sns.displot(df, x=column, kde=True)
```

Распределение длины лепестков ирисов с помощью violin плота.

```python
sns.violinplot(y=df['petal length'])
#https://seaborn.pydata.org/generated/seaborn.violinplot.html
```

Ещё один важный тип графиков -- pairplot, отражающий зависимость пар переменных друг от друга, а также распределение каждой из переменных.

```python
sns.pairplot(data=df, hue='species')
#https://seaborn.pydata.org/generated/seaborn.pairplot.html -- correlation between petal width and lenghth visible
```
